```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from lib import compare_imgs_with_figsize
compare_imgs = compare_imgs_with_figsize(figsize=(10, 10))
```

```python
img = cv2.imread('assets/x-ray.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
kernel_size = 5
kernel = np.ones((kernel_size, kernel_size), np.uint8)

dilation = cv2.dilate(img, kernel, iterations=1)

erosion = cv2.erode(img, kernel, iterations=1)

boundary = cv2.subtract(dilation, erosion)

compare_imgs(img, 'Original', dilation, 'Dilation', erosion, 'Erosion', boundary, 'Boundary', grid_shape=(2, 2), figsize=(10, 8))
```

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
compare_imgs(img, 'Original', opening, 'Opening', closing, 'Closing', grid_shape=(1, 3), figsize=(15, 10))
```

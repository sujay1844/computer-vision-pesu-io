```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
bw_img = cv2.imread('./lena.png')
bw_img = cv2.cvtColor(bw_img, cv2.COLOR_BGR2GRAY)
```

```python
bw_img_output = cv2.equalizeHist(bw_img)

img_hist = cv2.calcHist([bw_img], [0], None, [256], [0, 256])
img_output_hist = cv2.calcHist([bw_img_output], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.imshow(bw_img, cmap='gray')

plt.subplot(2, 2, 2)
plt.plot(img_hist)

plt.subplot(2, 2, 3)
plt.imshow(bw_img_output, cmap='gray')

plt.subplot(2, 2, 4)
plt.plot(img_output_hist)
```

```python
color_img = cv2.imread('./lena.png')
color_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)
```

```python
R_channel, G_channel, B_channel = cv2.split(color_img)

R_channel_eq = cv2.equalizeHist(R_channel)
G_channel_eq = cv2.equalizeHist(G_channel)
B_channel_eq = cv2.equalizeHist(B_channel)

color_img_eq = cv2.merge((R_channel_eq, G_channel_eq, B_channel_eq))
```

```python
R_channel_hist = cv2.calcHist([R_channel], [0], None, [256], [0, 256])
G_channel_hist = cv2.calcHist([G_channel], [0], None, [256], [0, 256])
B_channel_hist = cv2.calcHist([B_channel], [0], None, [256], [0, 256])

R_channel_eq_hist = cv2.calcHist([R_channel_eq], [0], None, [256], [0, 256])
G_channel_eq_hist = cv2.calcHist([G_channel_eq], [0], None, [256], [0, 256])
B_channel_eq_hist = cv2.calcHist([B_channel_eq], [0], None, [256], [0, 256])

plt.figure(figsize=(30, 15))

plt.subplot(2, 2, 1)
plt.imshow(color_img)

plt.subplot(2, 2, 2)
plt.plot(R_channel_hist, color='r')
plt.plot(G_channel_hist, color='g')
plt.plot(B_channel_hist, color='b')
plt.legend(['R', 'G', 'B'])

plt.subplot(2, 2, 3)
plt.imshow(color_img_eq)


plt.subplot(2, 2, 4)
plt.plot(R_channel_eq_hist, color='r')
plt.plot(G_channel_eq_hist, color='g')
plt.plot(B_channel_eq_hist, color='b')
plt.legend(['Red', 'Green', 'Blue'])

```
